```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection  import train_test_split

```

Data Collection and Processing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
breast_cancer_dataset
```

```python
df  = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
df.head()
```

Adding the target column to our dataset

```python
df["target"] = breast_cancer_dataset.target
df.tail()
```

Separating the Features and label

```python
x = df.drop(columns = "target", axis = 1)
y = df["target"]
```

Splitting the data into training set and test set



```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

Building the Neural NetworkDeepNeuralNetwork.png

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

Setting up the layers of Neural Network

```python
model = keras.Sequential([
                        keras.layers.Flatten(input_shape = (30,)),
                        keras.layers.Dense(20,activation="relu"),
                        keras.layers.Dense(2,activation="sigmoid")
])
```

```python
# Compiling the Neural Network
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

Training the neural network

```python
history = model.fit(x_train,y_train,validation_split = 0.1,epochs=10)
```

Visualizing Accuracy and loss

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["training data","validation data"],loc="lower right")
plt.show()
```

```python
 plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["training data","validation data"],loc="upper right")
plt.show()
```

```python
# Accuracy of the model on test data
loss, accuracy  = model.evaluate(x_test,y_test)

accuracy
```

```python
y_pred = model.predict(x_test)
y_pred
# neural network gives probability of whether this datapoint belogns to class 0 or 1 insted of ml models that directly give labels
```

Converting the prediction probability to class labels

```python
# argmax functions retruns the index of maximum no in a list
# also we are using for-loop to iterate through all of y_pred list

y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
